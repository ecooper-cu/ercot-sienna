```julia
using PowerSystems
using PowerSimulations
using HydroPowerSimulations
const PSI = PowerSimulations
using PowerSystemCaseBuilder
using Dates
using HiGHS
using TimeSeries
```

```julia
sys = build_system(PSISystems, "modified_RTS_GMLC_DA_sys")
```

```julia
components = [Area, FixedAdmittance, HydroDispatch, PowerLoad, RenewableDispatch, RenewableNonDispatch, VariableReserve]

for c in components
    for g in get_components(x -> has_time_series(x), c, sys)
        remove_time_series!(sys, DeterministicSingleTimeSeries, g, "max_active_power")
    end
end
```

```julia
transform_single_time_series!(
           sys,
           Dates.Hour(1), # horizon
           Dates.Minute(30), # interval
       );
```
